```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
import scipy.fftpack as sf
import scipy.signal as signal
```

```python
Fs=20
t=4
n=np.arange(0,t,1/Fs)
f=2
x=np.sin(2*pi*f*n)
```

```python
plt.plot(n,x)
```

```python
y=np.random.normal(0,0.8,len(x))
```

```python
plt.hist(y)
```

```python
x=x+y
```

```python
plt.plot(n,x)
```

```python
#spectral analysis
x_f=abs(sf.fft(x))
l=len(x)
```

```python
fr= (Fs/2)*np.linspace(0,1,int(l/2))
```

```python
xl_m=(1/2)*abs(x_f[0:np.size(fr)])
```

```python
plt.plot(fr,xl_m)
```

```python
fc=np.array([1.8,2.5])
o=4
```

```python
wc=2*fc/Fs
```

```python
[b,a]=signal.butter(o,wc,btype='bandpass')
```

```python
[W,h]=signal.freqz(b,a,worN=1024)
```

```python
W=Fs*W/(2*pi)
```

```python
plt.plot(W,20*np.log10(h))
```

```python
x_filt=signal.lfilter(b,a,x)
```

```python
plt.plot(n,x_filt)
```
